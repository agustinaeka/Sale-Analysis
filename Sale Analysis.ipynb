# import necesary libraries

```python
import pandas as pd
import os
```

### task1:merge the 12 months of sales data into a single csv file

```python


files = [file for file in os.listdir("./Sales_Data/")]

all_months_data = pd.DataFrame()

for file in files:
    data = pd.read_csv("Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data, data  ])
    
all_months_data.to_csv("all_data.csv")
```

### read the updated dataframe

# Clean up the data!

### Drop an NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
```

```python
nan_df
```

```python
all_data = all_data.dropna(how="any")
```

```python
all_data
```

### find "Or" and delete it

```python
or_data = all_data[all_data['Order Date'].str[0:2] == "Or"]
```

```python
or_data
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != "Or"]
```

# Augment data with additional Column

### task 2: Add month column

```python
all_data["month"] = all_data.loc[:, "Order Date"].str[0:2]
```

```python
# change str to int
all_data["month"] = all_data.loc[:, "Order Date"].str[0:2].astype("int32")
```

```python
all_data
```

## Question 1: What was the best month for sale ? How much was earned that month?

### task 3: Add sale column

```python
#convert column to exact type
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data["Price Each"])
```

```python
all_data["sale"] = all_data["Quantity Ordered"] * all_data["Price Each"]
```

```python
all_data
```

```python
results = all_data.groupby('month').sum()
```

```python
results
```

```python
import matplotlib.pyplot as plt
import numpy as np

month = np.arange(1,13)
```

```python
plt.bar(month, results["sale"])
plt.xticks(month)
plt.xlabel("Month")
plt.ylabel("Sales in USD")
plt.show()
```

### task 4: add city column

```python
def get_city(x):
    return x.split(",")[1]

def get_state(x):
    return x.split(",")[2].split(" ")[1]

all_data["city"] = all_data["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_state(x)}) ")
```

```python
all_data
```

```python
results = all_data.groupby(all_data["city"]).sum()
```

```python
results
```

```python
# city does not in correct order
city = all_data["city"].unique()

#city in correct order
cities = [city for city, df in all_data.groupby("city")]

plt.bar(cities, results["sale"])
plt.xticks(city, rotation="vertical", size=8)
plt.xlabel("City")
plt.ylabel("Number Of Sale")
plt.show()
```
